```python
from pathlib import Path

BASE_DIR = Path.cwd().resolve().parents[1]

#print("Project root:", BASE_DIR)
```

```python
path=BASE_DIR / "data"/ "APS"
```

```python
st_accept = "text/html" # говорим веб-серверу, 
                        # что хотим получить html
# имитируем подключение через браузер Mozilla на macOS
st_useragent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 12_3_1) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/15.4 Safari/605.1.15"
# формируем хеш заголовков
headers = {
   "Accept": st_accept,
   "User-Agent": st_useragent
}
```

```python
# импортируем модуль
import requests

link="https://www.apsnet.org/edcenter/resources/commonnames/Pages/default.aspx"


# отправляем запрос с заголовками по нужному адресу
req = requests.get(link, headers)
# считываем текст HTML-документа
src = req.text
print(src)
```

```python
from bs4 import BeautifulSoup

# инициализируем html-код страницы 
soup = BeautifulSoup(src, 'html.parser')
# считываем заголовок страницы
title = soup.title.string

#что-то не так со структурой повторяются элементы

for tag in soup.find_all(True):  # True находит все теги
    print(f'Tag: {tag.name}, Text: {tag.get_text(strip=True)}')
```

```python
plants=soup.find_all('li', class_='dfwp-item')

print(plants)
```

```python
plantsLinks={}
for tag in plants:
    plantsLinks[tag.get_text(strip=True).replace("Diseases of ","")]=tag.find('a').get('href')
```

```python
import json


json.dump(plantsLinks, open(path / "APSplants.txt",'w'))
```
